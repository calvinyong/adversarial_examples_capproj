```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## VGG like

```python
df = pd.read_csv('cifar10_vgglike_log.csv')
df.epoch += 1
```

```python
df.head()
```

```python
f, ax = plt.subplots(figsize=(13, 6))
ax = sns.lineplot(data=df[['acc', 'val_acc']], dashes=False)
ax.set_xlabel('epoch')
ax.set_ylabel('accuracy')
```

```python
f, ax = plt.subplots(figsize=(13, 6))
ax = sns.lineplot(data=df[['loss', 'val_loss']], dashes=False)
ax.set_xlabel('epoch')
ax.set_ylabel('loss')
```

```python
#df.iloc[[78, 84, 87, 111, 175]] original labels
df.iloc[[77, 83, 86, 110, 174]]
```
